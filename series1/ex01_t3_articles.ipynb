```python

import networkx as nx #for graphs
import matplotlib.pyplot as plt #for plotting
import numpy as np
import pandas as pd
```

```python
#LOAD ARTICLES

articles = pd.read_csv(
    "../DS/wikispeedia_paths-and-graph/articles.tsv",
    sep="\t",
    header=None,
    comment="#"
)

articles = articles.dropna()
articles_list = articles[0].tolist()

G=nx.from_pandas_adjacency

print("Total articles:", len(articles_list))
print(articles_list)
print(articles)


#LOAD LINKS
links = pd.read_csv(
    "../DS/wikispeedia_paths-and-graph/links.tsv",
    sep="\t",
    header=None,
    names=["source", "target"],
    comment="#"
)

links = links.dropna()
print("Total links:", len(links))
```

```python
#build graph from links

G=nx.from_pandas_edgelist(links, source="source", target="target",create_using=nx.DiGraph())
print(list(G.nodes())[:10]) #shows if we need to clean data, in our case we do
print(type(list(G.nodes())[0])) #we want class str here
#links are in the form of src and target ids, that are strings
```

```python
#find articles that dont appear in graph -> disconnected from the network
#A set is an unordered collection with no duplicate elements
nodes = set(G.nodes())
all_articles = set(articles_list)

missing_nodes=all_articles - nodes
print("Articles with no links", missing_nodes)
```

```python
#articles that have 0 degrees
isolated_nodes = list(nx.isolates(G))
print("Number of isolated nodes:", len(isolated_nodes))

in_degrees = dict(G.in_degree())
out_degrees = dict(G.out_degree())
print("In-degrees:", in_degrees)

for node, deg in in_degrees.items():
    if deg == 0:
        print(node) #nodes w. no in-degrees
        

#same thing diff notation
#zero_in_nodes = [n for n, d in G.in_degree() if d == 0] 
```
